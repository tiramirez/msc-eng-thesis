```python
import numpy as np
import pandas as pd
import csv
import os
```

## Actualizar OD

```python
os.chdir('./../00 Insumos/05 Object Detection')
os.listdir()
```

```python
with open('od_class_dict.csv') as file:
    reader = csv.reader(file)
    next(reader) 
    class_dict = {int(rows[0]):rows[1] for rows in reader}
#     class_dict.pop('key')
    
od = pd.read_csv('output_object_detection_stgo.csv')


od['class'] = np.vectorize(class_dict.get)(od['class'])
od['aux_val'] = 1
# od.head()


## Agrupar: Contar cada objeto por imagen
Pivot = pd.DataFrame(pd.pivot_table(od,
                                    index=['img_id'], 
                                    columns=['class'], 
                                    values=['aux_val'], 
                                    aggfunc='count').sum()).sort_values([0],ascending=False)
Pivot = Pivot.reset_index().drop(columns=['level_0'])

## Seleccionar top 15 objetos
top15_objects = Pivot['class'].head(20).tolist()
# Pivot.head(15)
# for i in Pivot['class'].values:
#     print(i)

## 
od_filtered = od[(od['class'].isin(top15_objects))]
od_features = pd.pivot_table(od_filtered,
               columns=['class'],
               index=['img_id'],
               values=['aux_val'],
              aggfunc='sum').fillna(0)
od_features.columns = [ 'od_' + i for i in od_features.columns.get_level_values(1) ]
od_features.index = pd.Index(od_features.index, name='ID')
```
