```python
# OD_veh_der = DefineVariable('OD_veh_der', 
#                             OD_cars_der + OD_bus_der + OD_motorcycle_der + OD_truck_der, database)
# OD_veh_izq = DefineVariable('OD_veh_izq', 
#                             OD_cars_izq + OD_bus_izq + OD_motorcycle_izq + OD_truck_izq, database)

B_mean_h = Beta('LLF: mean_h', 0, -100, 100, 0)
B_mean_l = Beta('LLF: mean_l', 0, -100, 100, 0)
B_mean_s = Beta('LLF: mean_s', 0, -100, 100, 0)
B_mean_bn = Beta('LLF: mean_bn', 0, -100, 100, 1)  # NS
B_stdev_h = Beta('LLF: stdev_h', 0, -100, 100, 0)
B_stdev_l = Beta('LLF: stdev_l', 0, -100, 100, 0)
B_stdev_s = Beta('LLF: stdev_s', 0, -100, 100, 0)
B_stdev_bn = Beta('LLF: stdev_bn', 0, -100, 100, 1)  # NS

B_edges = Beta('LLF: edges', 0, -100, 100, 0)
B_blobs = Beta('LLF: blobs', 0, -100, 100, 1)


B_Bicyclist_N = Beta('HLF N: Bicyclist', 0, -100, 100, 0) 
B_Building_N = Beta('HLF N: Building', 0, -100, 100, 0)
B_Car_N = Beta('HLF N: Car', 0, -100, 100, 0) 
B_Fence_N = Beta('HLF N: Fence', 0, -100, 100, 0) 
B_Pavement_N = Beta('HLF N: Pavement', 0, -100, 100, 0)  # NS 0
B_Pedestrian_N = Beta('HLF N: Pedestrian', 0, -100, 100, 0)
B_Pole_N = Beta('HLF N: Pole', 0, -100, 100, 0) 
B_Road_N = Beta('HLF N: Road', 0, -100, 100, 0)  # NS 0
B_SignSymbol_N = Beta('HLF N: SignSymbol', 0, -100, 100, 0) 
B_Sky_N = Beta('HLF N: Sky', 0, -100, 100, 0)
B_Tree_N = Beta('HLF N: Tree', 0, -100, 100, 1)

B_Bicyclist_S = Beta('HLF S: Bicyclist', 0, -100, 100, 0) 
B_Building_S = Beta('HLF S: Building', 0, -100, 100, 0) 
B_Car_S = Beta('HLF S: Car', 0, -100, 100, 0)
B_Fence_S = Beta('HLF S: Fence', 0, -100, 100, 0) 
B_Pavement_S = Beta('HLF S: Pavement', 0, -100, 100, 0)
B_Pedestrian_S = Beta('HLF S: Pedestrian', 0, -100, 100, 0)
B_Pole_S = Beta('HLF S: Pole', 0, -100, 100, 0)
B_Road_S = Beta('HLF S: Road', 0, -100, 100, 0)
B_SignSymbol_S = Beta('HLF S: SignSymbol', 0, -100, 100, 0)
B_Sky_S = Beta('HLF S: Sky', 0, -100, 100, 0) # NS 0 
B_Tree_S = Beta('HLF S: Tree', 0, -100, 100, 1) 


D_bench = Beta('OD: bench', 0, -100, 100, 0)
D_boat = Beta('OD: boat', 0, -100, 100, 0)
D_umbrella = Beta('OD: umbrella', 0, -100, 100, 0)
D_chair = Beta('OD: chair', 0, -100, 100, 0) ## SI VA

D_train = Beta('OD: train', 0, -100, 100, 0)## SI VA
D_potted_plant = Beta('OD: potted_plant', 0, -100, 100, 0)
D_stop_sign = Beta('OD: stop_sign', 0, -100, 100, 0)
D_traffic_light = Beta('OD: traffic_light', 0, -100, 100, 1)

D_bus = Beta('OD: bus', 0, -100, 100, 0)
D_motorcycle = Beta('OD: motorcycle', 0, -100, 100, 0)
D_bike = Beta('OD: bike BC', 0, -100, 100, 0)

# D_person = Beta('OD: person', 0, -100, 100, 1)
# D_car = Beta('OD: car', 0, -100, 100, 1)

D_fire_hydrant = Beta('OD: fire_hydrant', 0, -100, 100, 0)## SI VA
D_truck = Beta('OD: truck', 0, -100, 100, 0)## SI VA

D_car_boxcox = Beta('OD: car BC', 0, -100, 100, 0)## SI VA
LAMBDA_car = Beta('LAMBDA: car',1.0,0.0001,5,1)## SI VA

D_person_boxcox = Beta('OD: person BC', 0, -100, 100, 0)## SI VA
LAMBDA_person = Beta('LAMBDA: person',1.0,0.0001,5,1)## SI VA

D_bike_boxcox = Beta('OD: bike BC', 0, -100, 100, 0)## SI VA
LAMBDA_bike = Beta('LAMBDA: bike',1.0,0.0001,5,1)## SI VA

D_light_boxcox = Beta('OD: light BC', 0, -100, 100, 0)## SI VA
LAMBDA_light = Beta('LAMBDA: light',1.0,0.0001,5,1)## SI VA



# D_car_01 = Beta('OD: car 1', 0, -100, 100, 0)
# D_car_02 = Beta('OD: car 2', 0, -100, 100, 0)
# D_car_03 = Beta('OD: car 3', 0, -100, 100, 0)
# D_car_04 = Beta('OD: car 4', 0, -100, 100, 0)
# D_car_05 = Beta('OD: car 5', 0, -100, 100, 0)
# D_car_06 = Beta('OD: car 6-20', 0, -100, 100, 0)


tau1	 = Beta('tau1',-1,None,0,0)
delta2	 = Beta('delta2',2,0,None,0)

tau2 = tau1 + delta2

U = B_mean_h * mean_h + \
     B_mean_l * mean_l + \
     B_mean_s * mean_s + \
     B_mean_bn * mean_bn + \
     B_stdev_h * stdev_h + \
     B_stdev_l * stdev_l + \
     B_stdev_s * stdev_s + \
     B_stdev_bn * stdev_bn + \
     B_edges * edges + \
     B_blobs * blobs + \
     B_Bicyclist_N * N_Bicyclist + \
     B_Building_N * N_Building + \
     B_Car_N * N_Car + \
     B_Fence_N * N_Fence + \
     B_Pavement_N * N_Pavement + \
     B_Pedestrian_N * N_Pedestrian + \
     B_Pole_N * N_Pole + \
     B_Road_N * N_Road + \
     B_SignSymbol_N * N_SignSymbol + \
     B_Sky_N * N_Sky + \
     B_Tree_N * N_Tree + \
     B_Bicyclist_S * S_Bicyclist + \
     B_Building_S * S_Building + \
     B_Car_S * S_Car + \
     B_Fence_S * S_Fence + \
     B_Pavement_S * S_Pavement + \
     B_Pedestrian_S * S_Pedestrian + \
     B_Pole_S * S_Pole + \
     B_Road_S * S_Road + \
     B_SignSymbol_S * S_SignSymbol + \
     B_Sky_S * S_Sky + \
     B_Tree_S * S_Tree + \
     D_bench * OD_bench + \
     D_chair * OD_chair + \
     D_fire_hydrant * OD_fire_hydrant + \
     D_potted_plant * OD_potted_plant + \
     D_stop_sign * OD_stop_sign + \
     D_train * OD_train + \
     D_umbrella * OD_umbrella + \
     D_car_boxcox * models.boxcox(OD_car_der,LAMBDA_car) - \
     D_car_boxcox * models.boxcox(OD_car_izq,LAMBDA_car ) + \
     D_person_boxcox * models.boxcox(OD_person_der,LAMBDA_person) - \
     D_person_boxcox * models.boxcox(OD_person_izq,LAMBDA_person) + \
     D_truck * (OD_truck_der - OD_truck_izq) + \
     D_bike_boxcox * models.boxcox(OD_bicycle_der,LAMBDA_bike) - \
     D_bike_boxcox * models.boxcox(OD_bicycle_izq,LAMBDA_bike) + \
     D_light_boxcox * models.boxcox(OD_traffic_light_der,LAMBDA_light) - \
     D_light_boxcox * models.boxcox(OD_traffic_light_izq,LAMBDA_light)

#      D_car_01 * ( OD_Car_01_der - OD_Car_01_izq ) + \
#      D_car_02 * ( OD_Car_02_der - OD_Car_02_izq ) + \
#      D_car_03 * ( OD_Car_03_der - OD_Car_03_izq ) + \
#      D_car_04 * ( OD_Car_04_der - OD_Car_04_izq ) + \
#      D_car_05 * ( OD_Car_05_der - OD_Car_05_izq ) + \
#      D_car_06 * ( OD_Car_06_der - OD_Car_06_izq )


P1 = 1-dist.logisticcdf(U-tau1)
P2 = dist.logisticcdf(U-tau2)
Pe = dist.logisticcdf(U-tau1)- dist.logisticcdf(U-tau2)

ChoiceProba = {
    1: P1,
    3: Pe,
    2: P2}

# bioNormalCdf
# logisticcdf

logprob = log(Elem(ChoiceProba,choice))
formulas = {'loglike':logprob,
           'Prob 1':P1,
           'Prob =':Pe,
           'Prob 2':P2,
           'Utilidad':U}

biogeme = bio.BIOGEME(database,formulas,numberOfThreads=8)

biogeme.modelName = "LOG - safety_Wekun~base"

## algorithms
results = biogeme.estimate(algorithm=None)
R = results.getEstimatedParameters()

B = pd.concat([B,R['Value']],axis=1, sort=False)
T = pd.concat([T,R['t-test']],axis=1, sort=False)
```
